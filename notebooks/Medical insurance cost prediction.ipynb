## Project: Medical Insurance Cost Prediction Using Machine Learning (Python)

### Problem Statement

Insurance companies need accurate methods to estimate medical insurance charges based on an individual’s health and lifestyle factors. Traditional manual estimation can be inaccurate and slow. This project aims to build a machine-learning model that predicts the medical insurance cost for a person using demographic and health-related attributes.

### Objective

- To develop a predictive model that estimates insurance charges using features such as age, BMI, sex, children, smoking status, and region.

- To analyze which factors influence medical insurance cost the most.

- To build an end-to-end ML pipeline including data preprocessing, model training, testing, and evaluation.

### Data Pre-Processing - Encoding the Catergorical Features 

```python
#starting 5 rows of the data frame 

df.head()
```

```python
#encoding sex column
df.replace({"sex":{"male":0, "female":1}},inplace=True)

#encoding smoker column
df.replace({"smoker":{"yes":0, "no":1}},inplace=True)

#encoding region column
df.replace({"region":{"southeast":0, "southwest":1, "northeast":2, "northwest":3}},inplace=True)
```

```python
#starting 5 rows of the data frame 

df.head() 
```

```python
#removing the unwanted things from the column names 

df.columns = df.columns.str.strip()
df.head()
```

### Splitting the Features and Target

```python
X = df.drop(columns="charges",axis=1)
Y = df["charges"]
```

```python
print(X)
```

```python
print(Y)
```

### Splitting the data into Training and Testing

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

### Model Training

```python
#loading the linear regression model 

model = LinearRegression()
model.fit(X_train,Y_train)
```

### Model Evaluation

```python
#predicting the result using the trained model 

y_pred = model.predict(X_test)
```

```python
print("R2:", metrics.r2_score(Y_test, y_pred))
print("MAE:", metrics.mean_absolute_error(Y_test, y_pred))
print("MSE:", metrics.mean_squared_error(Y_test, y_pred))
```

### Building A Predictive System

```python
input_data = (31,1,25.74,0,1,0)

#changing input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the array
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)
print("The insurance cost is USD ",prediction[0])
```

### Conclusion 

**Key Insights**
- Smokers show 3–4× higher insurance charges.
